```python
import torch
import torch.nn as nn
import numpy as np
from matplotlib import pyplot as plt
```

```python
n=50
x=torch.randn(n,1)
y=x+torch.randn(n,1)/2
plt.plot(x,y,'o')
plt.show()
```

```python
ann_reg=nn.Sequential(
    nn.Linear(1,1),
    nn.ReLU(),
    nn.Linear(1,1)
)
lr=0.05
optimizer=torch.optim.SGD(ann_reg.parameters(),lr=lr)
loss_fun=nn.MSELoss()
```

```python
num_epochs=500
losses=torch.zeros(num_epochs)
for epochi in range(num_epochs):
  yhat=ann_reg(x)
  loss=loss_fun(yhat,y)
  losses[epochi]=loss

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
final_prediction=ann_reg(x)
final_loss=(final_prediction-y).pow(2).mean()
final_loss
```

```python
plt.plot(losses.detach(),'o')
plt.plot(num_epochs,final_loss.detach(),'ro')
plt.plot()
```
