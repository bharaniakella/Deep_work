```python
import torch
import torch.nn as nn
from matplotlib import pyplot as plt
import numpy as np
```

```python
# parameters
nPerClust = 100
blur = 1

A = [  1, 1 ]
B = [  5, 1 ]

# generate data
a = [ A[0]+np.random.randn(nPerClust)*blur , A[1]+np.random.randn(nPerClust)*blur ]
b = [ B[0]+np.random.randn(nPerClust)*blur , B[1]+np.random.randn(nPerClust)*blur ]

# true labels
labels_np = np.vstack((np.zeros((nPerClust,1)),np.ones((nPerClust,1))))

# concatanate into a matrix
data_np = np.hstack((a,b)).T

# convert to a pytorch tensor
data = torch.tensor(data_np).float()
labels = torch.tensor(labels_np).float()

# show the data
fig = plt.figure(figsize=(5,5))
plt.plot(data[np.where(labels==0)[0],0],data[np.where(labels==0)[0],1],'bs')
plt.plot(data[np.where(labels==1)[0],0],data[np.where(labels==1)[0],1],'ko')
plt.title('The qwerties!')
plt.xlabel('qwerty dimension 1')
plt.ylabel('qwerty dimension 2')
plt.show()
```

```python
def create_model(lr):
  ann_classify=nn.Sequential(
      nn.Linear(2,16),
      nn.ReLU(),
      nn.Linear(16,32),
      nn.ReLU(),
      nn.Linear(32,1),
      nn.Sigmoid()
  )
  loss_fun=nn.BCELoss()
  optimizer=torch.optim.SGD(ann_classify.parameters(),lr)
  return loss_fun,optimizer,ann_classify
```

```python
num_epochs=1000
losses=torch.zeros(num_epochs)
def train_model():
  for epochi in range(num_epochs):
    yhat=ann_classify(data)
    loss=loss_fun(yhat,labels)
    losses[epochi]=loss

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

  predictions=ann_classify(data)
  final_acc=torch.mean(((predictions>0.5)==labels).float())
  return losses,predictions,final_acc
```

```python
loss_fun,optimizer,ann_classify=create_model(0.01)
losses,predictions,final_acc=train_model()
```

```python
all_lr=torch.linspace(0.001,0.1,40)
accByLR=[]
all_losses=torch.zeros(len(all_lr),num_epochs)
for i,lr in enumerate(all_lr):
  loss_fun,optimizer,ann_classify=create_model(0.01)
  losses,predictions,final_acc=train_model()
  accByLR.append(final_acc)
  all_losses[i,:]=losses
```

```python
plt.plot(all_lr,accByLR,'s-')
plt.show()
```
